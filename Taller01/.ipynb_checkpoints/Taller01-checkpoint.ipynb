### Caso 1. Extactitud de esquemas de diferencias finitas

<p align="justify"> Utilice esquemas de diferencias finitas uniformes de 1er, 2do, y 4to orden de exactitud, para determinar los valores de primera y segunda derivada de la función mostrada en ecuación \ref{campana}</p>

\begin{equation}
    f(x) = \left[1.1cos(\kappa_1x)-1.3sin(\kappa_2x)\right]e^{-\kappa_0x^2}
    \label{campana}
\end{equation}

con $\kappa_0 = \frac{2}{3}$, $\kappa_1 = \frac{5}{3}$, $\kappa_2 = 3\kappa_1$

<p align="justify"> Use como valor de tamaño de malla máximo un valor constante igual a $\Delta x_{max}= \frac{\pi}{6}$, y complemente los esquemas propuestos con diferencias finitas asimétricas para la determinación de las derivadas en las fronteras</p>

<p align = "justify">La primera (ecuación \ref{firstD}) y segunda (ecuación \ref{secondD}) derivada de la ecuación \ref{campana} son, respectivamente: </p>

\begin{equation}
f^{'}(x) = \dfrac{\mathrm{e}^{-{\kappa}_0x^2}\cdot\left(26{\kappa}_0x\sin\left({\kappa}_2x\right)-13{\kappa}_2\cos\left({\kappa}_2x\right)-11{\kappa}_1\sin\left({\kappa}_1x\right)-22{\kappa}_0x\cos\left({\kappa}_1x\right)\right)}{10}
\label{firstD}
\end{equation}
    
\begin{multline}
f^{''}(x) = -\dfrac{\mathrm{e}^{-{\kappa}_0x^2}\cdot\left(\left(52{\kappa}_0^2x^2-13{\kappa}_2^2-26{\kappa}_0\right)\sin\left({\kappa}_2x\right)-52{\kappa}_0{\kappa}_2x\cos\left({\kappa}_2x\right)\right)}{10} \\
+\dfrac{\mathrm{e}^{-{\kappa}_0x^2}\cdot\left(-44{\kappa}_0{\kappa}_1x\sin\left({\kappa}_1x\right)+\left(-44{\kappa}_0^2x^2+11{\kappa}_1^2+22{\kappa}_0\right)\cos\left({\kappa}_1x\right)\right)}{10}
\label{secondD}
\end{multline}

<div class="alert alert-block alert-info"> <p align="justify"> <b>Punto 1.</b> Presentar la metodología de obtención de los siguientes esquemas de diferencias finitas para la primera derivada: </p></div>

<div class="alert alert-block alert-warning"> Obtener y emplear un esquema de segundo orden mediante cualquiera de las técnicas presentadas en clase (o la suya propia!) </div>

Vamos a obtener un esquema hacia adelante de orden dos para la primera derivada. Al despejar $f^{(1)}$ toda la serie de Taylor se divide por $\Delta x$. Así pues, para que el esquema sea de segundo orden, se debe conservar $f^{(1)}$

\begin{align}
 f_{i+1} &= f_{i} + \Delta x f^{(1)} + \dfrac{\Delta x^2}{2!} f^{(2)} +
 \dfrac{\Delta x^3}{3!} f^{(3)} 
\\
 f_{i+2} &= f_{i} + 2 \Delta x f^{(1)} + \dfrac{4\Delta x^2}{2!} f^{(2)} +
 \dfrac{8\Delta x^3}{3!} f^{(3)} \\
\end{align}

Multiplicamos cada ecuación por un peso, $a$ y $b$, y por comidad hacemos que el coeficiente que acompaña la primera derivada sea a:

\begin{equation}
 \frac{af_{i+1}}{\Delta x} = \frac{af_{i}}{\Delta x} + a f^{(1)} + a\frac{\Delta x}{2!} f^{(2)} + a\frac{\Delta x^2}{3!} f^{(3)} 
\label{ST1}
\end{equation}

\begin{equation}
 \frac{bf_{i+2}}{2\Delta x} = \frac{bf_{i}}{2\Delta x} + b f^{(1)} + b \dfrac{2\Delta x}{2!} f^{(2)} + b\dfrac{4\Delta x^2}{3!} f^{(3)}
 \label{ST2}
\end{equation}

De la expansión en series de Taylor se pueden obtener dos ecuaciones: una para la primera derivada, ecuación \ref{primDervida}, y otra para la segunda derivada, ecuación \ref{segDervida}:

\begin{equation}
    af^{(1)} + bf^{(1)} = f^{(1)}
    \\
    a + b = 1
    \label{primDervida}
\end{equation}

\begin{equation}
     a\dfrac{\Delta x}{2!} f^{(2)} + b \dfrac{2\Delta x}{2!} f^{(2)} = 0
     \\
     \dfrac{a}{2} + b  = 0
    \label{segDervida}
\end{equation}

De esta manera se puede plantear un sistema de dos ecuaciones y dos incógnitas:

$$
\left(\begin{array}{cc}
1 & 1  \\
1/2 & 1  
\end{array}\right)  \cdot  
\left(\begin{array}{cc}
a  \\
b  
\end{array}\right) = 
\left(\begin{array}{cc}
1 \\
0  
\end{array}\right)
$$

Al resolverla la matriz aumentanda usando la forma escalonada reducida por renglones, obtenemos que $a=2$ y $b=-1$. Si sumanos la serie de Taylor en \ref{ST1} y en \ref{ST2} reemplanzado los valores de $a$ y $b$:

\begin{equation}
    f^{(1)}_{i} = \frac{1}{\Delta x}\left[ 2f_{i+1} - \frac{f_{i+2}}{2} - 2f_{i} + \frac{f_{i}}{2} \right] + \frac{\Delta x^2}{3}f^{(3)}_{i} + H
\end{equation}

Usando MCM y truncando la serie en la tercera derivada encontramos el esquema FD de segundo orden para la primera derivada:

\begin{equation}
    f^{(1)}_{i} = \frac{1}{\Delta x}\left[\frac{-f_{i+2} + 4f_{i+1} - 3f_{i}}{2} \right] 
    \label{FD2O}
\end{equation}

Para obtener el esquema BD de segundo orden para la primera derivada, se puede emplear el esquema de la ecuación \ref{FD2O}. Únicamente hay "reflejar" el esquema, es decir, multiplicar el numerador por -1 y al igual que a los números dentro de los índices:

\begin{equation}
    f^{(1)}_{i} = \frac{1}{\Delta x}\left[\frac{f_{i-2} - 4f_{i-1} + 3f_{i}}{2} \right] 
    \label{BD2O}
\end{equation}

<div class="alert alert-block alert-warning"> Utilizar el esquema de cuarto orden compacto presentado en Lele (1992). </div>

\begin{equation}
\alpha\left(\frac{\partial \phi}{\partial x}\right)_{i+1}+\left(\frac{\partial \phi}{\partial x}\right)_i+\alpha\left(\frac{\partial \phi}{\partial x}\right)_{i-1}=\beta \frac{\phi_{i+1}-\phi_{i-1}}{2 \Delta x}+\gamma \frac{\phi_{i+2}-\phi_{i-2}}{4 \Delta x} .
\end{equation}

Para Padé-4 se usan las siguientes constantes: $\alpha=\frac{1}{4}$, $\beta = \frac{3}{2}$ y $\gamma = 0$. Se hacen los respectivos reemplazos y adicionalmente se multiplica la ecuación por 4:

\begin{equation}
\left(\frac{\partial \phi}{\partial x}\right)_{i+1}+4\left(\frac{\partial \phi}{\partial x}\right)_i+\left(\frac{\partial \phi}{\partial x}\right)_{i-1}=3\frac{\phi_{i+1}-\phi_{i-1}}{\Delta x}.
\end{equation}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
from icecream import ic
import time
```

```python
fun = lambda k0, k1, k2, x :(1.1*np.cos(k1*x) - 1.3*np.sin(k2*x))*np.exp(-k0*x**2)

Der1F = lambda k0, k1, k2, x : np.exp(-k0*x**2)*(26*k0*x*np.sin(k2*x) - \
                               13*k2*np.cos(k2*x) - 11*k1*np.sin(k1*x) - 22*k0*x*np.cos(k1*x))/10

Der2F = lambda k0, k1, k2, x : -np.exp(-k0*x**2)*((52*k0**2*x**2 - 13*k2**2 - 26*k0)*np.sin(k2*x) - 52*k0*k2*x*np.cos(k2*x) \
                                - 44*k0*k1*x*np.sin(k1*x) + (-44*k0**2*x**2 + 11*k1**2 + 22*k0)*np.cos(k1*x))/10
```

```python
def firtOrder(nodes, xDelta, precision):
        A1 = np.eye(nodes,nodes, k=0, dtype = np.dtype(precision))*-1 +\
            np.eye(nodes,nodes, k=1, dtype = np.dtype(precision))*1 
        A1[-1,-1] = 1; A1[-1,-2] = -1; A1 = A1*(1/xDelta); A1 = A1.astype(dtype = np.dtype(precision))
        return A1

def secondOrder(nodes, xDelta, precision):
        A2 = np.eye(nodes,nodes, k=1, dtype = np.dtype(precision))*1 +\
            np.eye(nodes,nodes, k=-1, dtype = np.dtype(precision))*-1  
        border = np.array([-3, 4, -1], dtype=np.dtype(precision))
        A2[0,:len(border)] = border
        A2[-1,-len(border):] = np.flip(border)*-1
        A2 = A2*(1/(2*xDelta)); A2 = A2.astype(dtype = np.dtype(precision))
        return A2

def fourthOrder(nodes, xDelta, precision):
        A4 = np.eye(nodes,nodes, k=0, dtype = np.dtype(precision))*4 +\
             np.eye(nodes,nodes, k= 1, dtype = np.dtype(precision))*1 +\
             np.eye(nodes,nodes, k=-1, dtype = np.dtype(precision))*1  
        A4[0,1], A4[-1,-2], A4[0,0], A4[-1,-1] = 0, 0, 1, 1 
        A4 = A4.astype(dtype = np.dtype(precision))
        A4_inv = np.linalg.inv(A4)

        B4 = np.eye(nodes, nodes, k=1, dtype = np.dtype(precision))*3 +\
             np.eye(nodes, nodes, k=-1, dtype = np.dtype(precision))*-3
        border = np.array([-25/12, 48/12,-36/12, 16/12, -3/12], dtype=np.dtype(precision))
        B4[0,:len(border)] = border
        B4[-1, -len(border):] = np.flip(border)*-1
        B4 = B4*(1/(xDelta))
        
        return A4_inv, B4

def numDer_orderN(nodes, precision, A, F, B=1, typeScheme="normal"):
        if typeScheme == "normal":
                derN_orderN = np.zeros(nodes, dtype = np.dtype(precision)); derN_orderN = np.reshape(derN_orderN, (nodes,1))
                derN_orderN = np.dot(A,F)
        elif typeScheme == "compact":
                derN_orderN = np.zeros(nodes, dtype = np.dtype(precision)); derN_orderN = np.reshape(derN_orderN, (nodes,1))
                derN_orderN = np.dot(np.dot(A,B),F)

        return derN_orderN
```

```python
def firstDerivative(precision, numbDeltas, k0, k1, k2, type = "firstDer"):

    xDelta = np.zeros((1,numbDeltas),  dtype = np.dtype(precision))
    xDelta = np. reshape(xDelta,(1, numbDeltas))

    if precision == "float16":
        for i in range(1,numbDeltas+1):
            xDelta[0,i-1] = np.float16(np.pi/2**(i+2))
    elif precision == "float32":
        for i in range(1,numbDeltas+1):
            xDelta[0,i-1] = np.float32(np.pi/2**(i+2))
    else:
        for i in range(1,numbDeltas+1):
            xDelta[0,i-1] = np.float64(np.pi/2**(i+2))
            
    Data = np.zeros((len(xDelta[0]), 3), dtype = np.dtype(precision))
    Data = np.reshape(Data, (len(xDelta[0]), 3))

    for j in range(len(xDelta[0])):
        if precision == "float16":
            nodes = np.int_(np.ceil(np.float16(np.pi*2)/xDelta[0,j])) + 1
        elif precision == "float32":
            nodes = np.int_(np.ceil(np.float32(np.pi*2)/xDelta[0,j])) + 1
        else:
            nodes = np.int_(np.ceil(np.float64(np.pi*2)/xDelta[0,j])) + 1

        xCoord = np.linspace(-np.pi, np.pi, nodes, dtype = np.dtype(precision))
        
        #Create the F column vector that contain the values of the function
        F = np.zeros(nodes, dtype = np.dtype(precision))
        F = fun(k0, k1, k2, xCoord); F = np.reshape(F, (nodes,1))

        if type == "firstDer":
            #First order
            A1 = firtOrder(nodes, xDelta[0,j], precision)
            #Second order
            A2 = secondOrder(nodes, xDelta[0,j], precision)
            #Fourth order
            A4_inv, B4 = fourthOrder(nodes, xDelta[0,j], precision)
            #Compute the derivative
            der1_order1 = numDer_orderN(nodes, precision, A1, F)
            der1_order2 = numDer_orderN(nodes, precision, A2, F)
            der1_order4 = numDer_orderN(nodes, precision, A4_inv, F, B4, typeScheme="compact")
            #Compute L2 error
            der1Analy = np.reshape(Der1F(k0, k1, k2, xCoord), (nodes,1))
            Data[j,0] = np.linalg.norm(der1Analy-der1_order1)
            Data[j,1] = np.linalg.norm(der1Analy-der1_order2)
            Data[j,2] = np.linalg.norm(der1Analy-der1_order4)
        elif type == "secondDer":
            #First order
            A1 = firtOrder(nodes, xDelta[0,j], precision)
            #Second order
            A2 = secondOrder(nodes, xDelta[0,j], precision)
            #Compute the derivative
            der2_order1 = numDer_orderN(nodes, precision, np.dot(A1,A1), F)
            der2_order2 = numDer_orderN(nodes, precision, np.dot(A2,A2), F)
            #Compute L2 error
            der2Analy = np.reshape(Der2F(k0, k1, k2, xCoord), (nodes,1))
            Data[j,0] = np.linalg.norm(der2Analy-der2_order1)
            Data[j,1] = np.linalg.norm(der2Analy-der2_order2)
            

            
    return xDelta, Data
```

```python
def plotter(xDelta1, Data1, xDelta2, Data2, numDeltas, precision, type = "firstDer"):
    Dict = {}
    Dict[0] = {'Data': Data1, 'xDelta': xDelta1}
    Dict[1] = {'Data': Data2, 'xDelta': xDelta2}
    letter = ["a", "b"]
    i = 0
    fig, ax = plt.subplots(1,2, figsize=(10, 5), sharex = True, sharey = True)
    for info in Dict:
        xStep = np.reshape(Dict[info]["xDelta"], (numDeltas,1))
        
        ax[i].set_yscale("log")
        ax[i].set_xscale("log")
        ax[i].plot(xStep, np.reshape(Dict[info]["Data"][:,0], (numDeltas, 1)), "--bo", label='$1^{er}$ orden')
        ax[i].plot(xStep, np.reshape(Dict[info]["Data"][:,1], (numDeltas, 1)), "--go", label='$2^{do}$ orden')
        if type == "firstDer":
            ax[i].plot(xStep, np.reshape(Dict[info]["Data"][:,2], (numDeltas, 1)), "--ko", label='$4^{to}$ orden')
            ax[i].set_title("$1^{er}$, $2^{do}$ y $4^{to}$ orden, precisión "+ precision[i])
            fig.suptitle('Análisis de errores y precisión de esquemas FDM  para primera derivada')
            ax[i].set_xlabel("Tamaño de la malla $\Delta x$ \nFigura 1" + letter[i] +\
                             ": $1^{er}$, $2^{do}$ y $4^{to}$ orden, precisión"  + precision[i])
        elif type == "secondDer":
            ax[i].set_title("$1^{er}$ y $2^{do}$ orden, precisión "+ precision[i])
            fig.suptitle('Análisis de errores y precisión de esquemas FDM  para segunda derivada')
            ax[i].set_xlabel("Tamaño de la malla $\Delta x$ \nFigura 2" + letter[i] +\
                             ": $1^{er}$ y $2^{do}$ orden, precisión "  + precision[i])

        
        ax[i].set_ylabel("Error $L_2$ ")
        ax[i].legend()
        ax[i].grid()
        i+=1
    return ax
    
```

```python
k0 = 2/3; k1 = 5/3; k2 = 3*k1
numDeltas = 10
precision = np.array(['float32', 'float64'])
xDelta1, Data1 = firstDerivative(precision[0], numDeltas, k0, k1, k2)
xDelta2, Data2 = firstDerivative(precision[1], numDeltas, k0, k1, k2)
```

<div align='justify'> Haciendo una primera inspección visual en la región estable de la <b>Figura 1a</b> y en la <b>Figura 1b</b>, se puede identificar cómo se comporta el error $L_2$ con el error de truncamiento del esquema numérico. Simplemente, a mayor exactitud del esquema (es decir, menor error de truncamiento), mayor es la disminución del error $L_2$ a medida que se usan tamaños de malla más finos. Curiosamente, la pendiente de dicha recta es aproximadamente $1/(\Delta x)^{orden}$. Esto tiene una relevancia significativa en cuanto a costo computacional. Entre mayor sea el orden del esquema, menor es la necesidad de usar mallas finas, es decir, existe la posibilidad de implementar mallas gruesas. ¿Qué ocasiona esto? Un número menor de divisiones en el dominio espacial, generando un menor número de nodos en el cálculo y por tanto, menor tamaño de las matrices usadas en el cálculo computacional. </div><br>

<div align='justify'> También se puede encontrar una relación entre el error $L_2$ con el error de redondeo a medida que se incrementa la precisión en los cálculos. En la <b>Figura 1a</b> se puede disitnguir una situación que se podría tildar de "anomala", y es que llega un punto en el cual el error $L_2$ no sigue decreciendo, sino que empieza a crecer. A esta región se le conoce como zona inestable. Básicamente, esto se debe a que el error de redondeo cobra mayor relevancia en tamaños de malla muy finos y sin importar que tanto se refine, el error $L_2$ seguirá aumentado. El primero que "muerde" el error de redondeo es el esquema cuarto orden, y luego le sigue el esquema de segundo orden.</div><br>

<div align='justify'>Sin embargo, hay una forma de "postergar" la zona inestable, y es aumentando la precisión en los cálculos, es decir, reservando mayor memoria del computador para almacenar y operar números. ### Caso 2. Transporte difusivo y decaimiento de contaminantes

\begin{equation}
\begin{aligned}
& \frac{\partial P}{\partial t}=\frac{\partial}{\partial x}\left[D(x) \frac{\partial P}{\partial x}\right]-u \frac{\partial P}{\partial x}-\tau P+S \\
& \frac{\partial Q}{\partial t}=\frac{\partial}{\partial x}\left[D(x) \frac{\partial Q}{\partial x}\right]-u \frac{\partial Q}{\partial x}+\tau P
\label{caso2}
\end{aligned}
\end{equation}


<div class="alert alert-block alert-info"> <p align="justify"> <b>Punto 1.</b> Generar expresiones adimensionales de las ecuaciones </p></div>
<br>

Para adimensionalizar el sistema de ecuaciones diferenciales parciales, primero se deben proponer variables adimensioanles:

$$
x^* = \frac{x}{L_c},~t^* = \frac{t}{t_c}, ~Q^* = \frac{Q}{c}, ~P^* = \frac{P}{c}
$$

Donde $L_c$ es la longitud total $L$ del dominio $x$, $c$ es la cantidad de un reactivo $P$ y $Q$ en $x=L$ y el tiempo característico es igual a $L_c/u$. Posteriormente se reemplazan las expresiones adimensionales en la ecuación \ref{caso2}:

\begin{equation}
\begin{aligned}
&\frac{\partial P^*}{\partial t^*}=\frac{1}{uL_c}\frac{\partial}{\partial x^*}\left[D(x) \frac{\partial P^*}{\partial x^*}\right]- \frac{\partial P^*}{\partial x^*} - \frac{\tau L_c}{u} P^* + \frac{L_c}{uc}S \\
&\frac{\partial Q^*}{\partial t^*}=\frac{1}{uL_c}\frac{\partial}{\partial x^*}\left[D(x) \frac{\partial Q^*}{\partial x^*}\right]- \frac{\partial Q^*}{\partial x^*} + \frac{\tau L_c}{u} P^*
\label{adimensional}
\end{aligned}
\end{equation}


<div class="alert alert-block alert-info"> <p align="justify"> <b>Punto 2.</b> Formular esquemas discretos para las expresiones adimensionales usando NuCFDS acopladas con
Crank-Nicolson</p></div>
<br>

$$
K_1=\frac{1}{u L_c} ; K_2=\frac{G L_c}{u} ; K_3=\frac{L_c}{u \cdot c} ; K_4=\frac{u \cdot L_c}{D}
$$

<br>

A continuación se presenta la obtención del esquema Crank-Nicolson para la ecuación del reactivo $Q$, ya que para $P$ se obtuvo en clase. Primero se plantea el promedio entre los términos de la ecuación en el tiempo $n+1$ y en el tiempo $n$ utilizando esquemas NUCFDS: 

$$
\begin{aligned}
\frac{Q_i^{n+1}-Q_i^n}{\Delta t}= & \frac{k_1}{2}\left\{\frac{D_{i+1 / 2} \frac{Q_{i+1}^{n+1}-Q_i^{n+1}}{x_{i+1}-x_i}-D_{i-1 / 2} \frac{Q_i^{n+1}-Q_{i-1}^{n+1}}{x_i-x_{i-1}}}{1 / 2\left[x_{i+1}-x_{i-1}\right]}\right\}-\frac{1}{2}\left\{\frac{Q_{i+1}^{n+1}-Q_{i-1}^{n+1}}{x_{i+1}-x_{i-1}}\right\}+\frac{k_2}{2} P_i^{n+1} \\
& \frac{k_1}{2}\left\{\frac{D_{i+1 / 2} \frac{Q_{i+1}^n-Q_i^n}{x_{i+1}-x_i}-D_{i-1 / 2} \frac{Q_i-Q_{i-1}^n}{x_i-x_{i-1}}}{1 / 2\left[x_{i+1}-x_{i-1}\right]}\right\}-\frac{1}{2}\left\{\frac{Q_{i+1}^n-Q_{i-1}^n}{x_{i+1}-x_{i-1}}\right\}+\frac{k_2}{2} P_i^n
\end{aligned}
$$
<br>
Se busca simplicar la expresión de tal manera que se puedan comprimir los coeficientes extensos en letras como $\alpha$, $\beta$, $\gamma$ y $\delta$:

$$
\begin{aligned}
Q_i^{n+1}=Q_i^n & +\frac{k_1 \Delta t}{2}\left\{\frac{D_{i+1 / 2}\left(Q_{i+1}^{n+1}-Q_i^{n+1}\right)\left(x_i-x_{i-1}\right)-D_{i-1 / 2}\left(Q_i^n-Q_{i-1}^n\right)\left(x_{i+1}-x_i\right)}{1 / 2\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}x_i\right)\left(x_i-x_{i-1}\right)}\right\} \\ 
& -\frac{\Delta t}{2\left(x_{i+1}-x_{i-1}\right)}\left(Q_{i+1}^{n+1}-Q_{i-1}^{n+1}\right) +\frac{k_2 \Delta t}{2} P^{n+1}  \\
& + \frac{k_1 \Delta t}{2}\left\{\frac{D_{i+1 / 2}\left(Q_{i+1}^n-Q_i^n\right)\left(x_{i-}-x_{i-1}\right)-D_{i-1 / 2}\left(Q_i^n-Q_{i-1}^n\right)\left(x_{i+1}-x_i\right)}{\frac{1}{2}\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}\right\} \\ 
& -\frac{\Delta t\left(Q_{i+1}^n-Q_{i-1}^n\right)}{2\left(x_{i+1}-x_{i-1}\right)} + \frac{k_2 \Delta t}{2} P^{n}
\end{aligned}
$$

<br>

$$
\begin{aligned}
Q_i^{n+1}= & Q_i^n+\frac{k_1 \Delta t D_{i+1 / 2}\left(x_i-x_{i-1}\right)}{\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}\left(Q_{i+1}^{n+1}-Q_i^{n+1}\right)-\frac{k_1 \Delta t D_{i-1 / 2}\left(x_{i+1}-x_i\right)}{\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}\left(Q_{i-1}^{n+1}-Q_{i-1}^{n+1}\right) \\
& -\frac{\Delta t}{2\left(x_{i+1}-x_{i-1}\right)}\left(Q_{i+1}^{n+1}-Q_{i-1}^{n+1}\right)+\frac{k_2 \Delta t}{2} P^{n+1} \\
& +\frac{k_1 \Delta t D_{i+1 / 2}\left(x_i-x_{i-1}\right)}{\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}\left(Q_{i+1}^{n}-Q_i^{n}\right)-\frac{k_1 \Delta t D_{i-1 / 2}\left(x_{i+1}-x_i\right)}{\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}\left(Q_{i-1}^{n}-Q_{i-1}^{n}\right) \\
& -\frac{\Delta t}{2\left(x_{i+1}-x_{i-1}\right)}\left(Q_{i+1}^{n}-Q_{i-1}^{n}\right)+\frac{k_2 \Delta t}{2} P^{n}
\end{aligned}
$$

<br>
Los coeficientes extensos se comprimen de la siguiente manera:

$$
\alpha = \frac{k_1 \Delta t D_{i+1 / 2}\left(x_i-x_{i-1}\right)}{\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}
$$

$$
\beta = \frac{k_1 \Delta t D_{i-1 / 2}\left(x_{i+1}-x_i\right)}{\left(x_{i+1}-x_{i-1}\right)\left(x_{i+1}-x_i\right)\left(x_i-x_{i-1}\right)}
$$

$$
\gamma = \frac{\Delta t}{2\left(x_{i+1}-x_{i-1}\right)}
$$

$$
\delta = \frac{k_2 \Delta t}{2}
$$

<br>
Si se dejan las incógnitas al lado izquierdo y las magnitudes conocidas al lado derecho se obtiene el esquema Crank-Nicolson con NUCFDS de segundo orden:

\begin{equation}
-(\beta+\gamma) Q_{i-1}^{n+1}+(1+\alpha+\beta) Q_i^{n+1}-(\alpha-\gamma) Q_{i+1}^{n+1}= 
(\beta+\gamma) Q_{i-1}^n+(1-\alpha-\beta) Q_i^n+(\alpha-\gamma) Q_{i+1}^n+\delta\left(P_i^n+P_i^{n+1}\right) 
\label{Qdiscre}
\end{equation}

<br>

Sin embargo, esta expresión sirve únicamente para los nodos internos. Es necesario discretizar la condición de frontera para $x=0$ y armar una expresión que sirva al momento de evaluar el nodo $i=1$.

$$

Se reemplaza la expresión para el nodo $i-1$ en la ecuación \ref{Qdiscre}:

\begin{aligned}

& +b_2 Q_i^n+c Q_{i+1}^n+\delta\left(p^{n+1}+p^n\right)
\end{aligned}
$$

```python
# START Definicion de funciones
def profileS(posX,xFrac,maxS,sigma):
    xMin = np.float64(posX[0])
    xMax = np.float64(posX[-1])
    totalL = xMax - xMin
    return maxS * np.exp(-0.5*(np.power((posX-xFrac*totalL)/sigma,2.0)))

def profileD(posX,minD):
    xMin = np.float64(posX[0])
    xMax = np.float64(posX[-1])
    totalL = xMax - xMin
    return minD + 0.05*(np.sin(2.5*posX*np.pi/totalL)+1.0)
# END Definicion de funciones

def main():
    # Lectura de datos entrada
    simData = pd.read_csv("simDataDifConvReacDual1D.dat")
    simData.to_numpy()
    simData.columns = simData.columns.str.strip()
    simData.set_index('Variable',inplace=True)
    domainL  = np.float64(simData.loc['domainL']['Value'])
    nNodesX  = np.int64(simData.loc['nNodesX']['Value'])
    meshExpRat  = np.float64(simData.loc['meshExpRat']['Value'])
    posSPeak  = np.float64(simData.loc['posSPeak']['Value'])
    maxS  = np.float64(simData.loc['maxS']['Value'])    
    sigmaS  = np.float64(simData.loc['sigmaS']['Value'])
    minDiffVal  = np.float64(simData.loc['minDiffVal']['Value'])
    print(simData)
    positionsX = np.zeros(int(nNodesX),dtype=np.float64)
    deltasX=np.zeros(int(nNodesX),dtype=np.float64)
    # Valores posicionales se calculan internamente
    if (np.abs(meshExpRat-1.0E0)<1.0E-10):
        startDeltaX = domainL/np.float64(nNodesX-1)
        deltasX[:] = startDeltaX
    else:
        startDeltaX = domainL*(meshExpRat - 1.0)/(np.power(meshExpRat,int(nNodesX)-1)-1.0)
        deltasX[0] = startDeltaX/meshExpRat
        for i in range(1,int(nNodesX)):
          deltasX[i] = np.power(meshExpRat,i-1)*startDeltaX
    
    for i in range(1,int(nNodesX)):
      positionsX[i] = positionsX[i-1] + deltasX[i]
    
    # Generamos perfiles y los guardamos en archivo,
    # luego los volvemos a adquirir con lectura de CSV en el programa principal.
    valsS = profileS(positionsX,posSPeak,maxS,sigmaS)
    #fig, ax = plt.subplots(1,2, figsize=(8, 5), sharex = True)
    #ax[0].plot(positionsX,valsS,'o-')
    #ax[0].grid("True")
    valsD = profileD(positionsX,minDiffVal)
    #ax[1].plot(positionsX,valsD,'o-')
    #ax[1].grid("True")

    # Escribiendo valores en archivo perfilesDS.dat
    perfilesDS = np.zeros((int(nNodesX),2))
    perfilesDS[:,0] = np.transpose(valsD)
    perfilesDS[:,1] = np.transpose(valsS)
    perfilesDSAsDF=pd.DataFrame(perfilesDS)
    perfilesDSAsDF.to_csv('./perfilesDS.dat',index=False,header=False,index_label=False)

if __name__ == "__main__":
    main()
```

Con esta configuración ocurre un fenómeno llamado <b>Oscilación numérica</b> que se puede apreciar justo al final del dominio espacial. Consiste en saltos u oscilaciones que no tienen sentido físico. Según [4], a pesar de que Crank-Nicolson es incondicionalmente estable, puede presentar este tipo fenómeno si existen discontinuidades en las condiciones iniciales o entre las condiciones iniciales y de frontera. ¿Algo de esto ocurre acá? Al refinar la malla al final del dominio, se está "acentuando" esta discontinuidad, es decir, en un cambio más pequeño de posición, se da un cambio de $0$ a $1$ en cantidad de reactivo. Y al parecer, esta oscilación numérica se propaga a lo largo del domino temporal.</div><br>

Esto ocasionó que las curvas de reactivo $P$ y $Q$ se "aplanaran" en medio del dominio. Esto tiene una razón física muy sencilla, y es que ahora el contaminante se está transportando en el río a mayor velocidad, quitando así capacidad de acumulación por término fuente. Adicionalmente quería comentar, aunque en este estudio no ocurrió, también pueden ocurrir oscilaciones numéricas (bien marcadas) cuando se discretiza el término advectivo en presencia de altas velocidades de transporte. Esto se debe a que la "información viaja muy rápido" y la malla no es capas de captarla. Esta situación puede ocurrir incluso con Crank-Nicolson, siendo un esquema indocionalmente estable. Para solucionar este problema, se recomienda utilizar un esquema que pueda recibir información desde aguas arriba, como un esquema hacia adelante o hacia atrás.</div><br>

### Caso 3. Transporte difusivo en dominio bi-dimensional

<div class="alert alert-block alert-info"> <p align="justify"> <b>Punto 1.</b> Generar expresiones adimensionales de las ecuaciones </p></div>
<br>

Primero se escogen las constantes características con las que se busca adimensionalizar:
<br>
$$
\begin{aligned}
& \delta t=\delta t^* \cdot \tau \\
& \delta x=\delta x^* L_c \\
& \delta y=\delta y^* L_c \\
& \delta u=\delta u^* u_c \\
& \delta v =\delta v^* u_c \\
\end{aligned}
$$
<br>
Al reemplazar cada una de estas constantes se obtiene, para la ecuación de la velociadad, la siguiente ecuación adimensional:
<br>

 $$
\frac{u_c}{\tau} \frac{\delta u^*}{\delta t^*}+\frac{u^* \cdot u_c^2}{L_c} \frac{\delta u^*}{\delta x^*}+\frac{v^* u_c^2}{L_c} \cdot \frac{\delta u^*}{\delta x^*}=S_x
$$
<br>
Al despejar el término temporal, tanto para el tranporte en $x$ y en $y$ se obtiene:
<br>

\begin{equation}
\begin{aligned}
& \frac{\delta u^*}{\delta t^*}+\frac{u_c \cdot \tau}{L_c} u^* \frac{\delta u^*}{\delta x^*}+\frac{u_c \cdot \tau}{L_c} v^* \frac{\delta u^*}{\delta y^*}=\frac{\tau}{u_c} S_x \\
& \frac{\delta v^*}{\delta t^*}+\frac{u_c \cdot \tau}{L_c} u^* \frac{\delta u^*}{\delta x^*}+\frac{u_c \cdot \tau}{L_c} v^* \frac{\delta v^*}{\delta y^*}=\frac{\tau}{u_c} S_y
\end{aligned}
\end{equation}
<br>
donde $k_1= \frac{u_c \cdot \tau}{L_c} $ y $k_2=\frac{\tau}{u_c}$

<div class="alert alert-block alert-info"> <p align="justify"> <b>Punto 2.</b> Formular esquemas discretos para las expresiones adimensionales, usando UCFDS de segundo orden. </p></div>
<br>

La ecuación de advección-reacción es no lineal debido a que los coeficientes contienen incógnitas, ¿Qué problema trae esto? Que en principio no se puede hacer un planteamiento matricial para resolver el sistema. Sin embargo, se puede hacer una linealización, evaluando las velocidades de los coeficiente no en el tiempo futuro, sino en el presente. La ventaja de esta técnica es que permite plantear la solución en forma matricial. La desventaja es que introduce un error adicional en el cálculo, porque esas velocidades deberían ser evaluadas en tiempo futuro y no en el tiempo presente. A continuación se presenta la dsicretización:

<br>

$$
\begin{array}{r}
\frac{u_{i, j}^{n+1}-u_{i, j}^n}{\Delta t}=-\frac{k_1}{2} u_{i, j}^n\left[\frac{u_{i+1}^{n+1}-u_{i-1}^{n+1}}{2 \Delta x}\right]-\frac{k_1}{2} v_{i, j}^n\left[\frac{u_{j+1}^{n+1}-u_{j-1}^{n+1}}{2 \Delta y}\right]+\frac{k_2}{2} s_x^{n+1} \\
\\
-\frac{k_1}{2} u_{i, j}^n\left[\frac{u_{i+1}^n-u_{i-1}^n}{2 \Delta x}\right]-\frac{k_1}{2} v_{i, j}^n\left[\frac{u_{j+1}^n-u_{j-1}^n}{2 \Delta y}\right]+\frac{k_2}{2} s_x^n
\end{array}
$$

<br>

$$
\begin{aligned}
u_{i, j}^{n+1}=u_{i, j}^n & =\frac{\Delta t k_1}{2} u_{i, j}^n\left[\frac{u_{i+1}^{n+1}-u_{i-1}^{n+1}}{2 \Delta x}\right]-\frac{\Delta t k_1}{2} v_{i, j}^n\left[\frac{u_{j+1}^{n+1}-u_{j-1}^{n+1}}{2 \Delta y}\right]+\frac{\Delta t k_2 S_x^{n+1}}{2} \\
& -\frac{\Delta t k_1}{2} u_{i, j}^n\left[\frac{u_{i+1}^n-u_{i-1}^n}{2 \Delta x}\right]-\frac{\Delta t k_1}{2} v_{i, j}^n\left[\frac{u_{j+1}^n-u_{j-1}^n}{2 \Delta y}\right]+\frac{\Delta t k_2}{2} S_x^n
\end{aligned}
$$

<br>

$$
\begin{aligned}
u_{i, j}^{n+1}=u_{i, j}^n & -\frac{\Delta t k_1 u_{i, j}^n}{4 \Delta x} u_{i+1}^{n+1}+\frac{\Delta t k_1 u_{i, j}^n}{4 \Delta x} u_{i-1}^{n+1}-\frac{\Delta t k_1 v_{i, j}^n}{4 \Delta y} u_{j+1}^{n+1}+\frac{\Delta t k_1 u_{i, j}^n}{4 \Delta y} u_{j-1}^{n+1}+\frac{\Delta t k_2 S_x^{n+1}}{2} \\
& -\frac{\Delta t k_1 u_{i, j}^n}{4 \Delta x} u_{i+1}^n+\frac{\Delta t k_1 u_{i, j}^n}{4 \Delta x} u_{i-1}^n-\frac{\Delta t k_1 v_{i, j}^n}{4 \Delta y} u_{j+1}^n+\frac{\Delta t k_1 v_{i, j}^n }{4 \Delta y}u_{j-1}^n+\frac{\Delta t k_2 S_x^n}{2}
\end{aligned}
$$
<br>
Con:
<br>
$$
\alpha =\frac{\Delta t k_1 u_{i,j}^{n}}{4\Delta x} ; \beta =\frac{\Delta t k_1v_{i,j}^{n}}{4\Delta y} ; \gamma =\frac{\Delta t k_2}{2}
$$
<br>
Se obtiene:
<br>
$$
\begin{aligned}
u_{i j}^{n+1}=u_{i, j}^n & -\alpha u_{i+1}^{n+1}+\alpha u_{i-1}^{n+1}-\beta u_{j+1}^{n+1}+\beta u_{j-1}^{n+1}+\gamma s_x^{n+1} \\
& -\alpha u_{i+1}^n+\alpha u_{i-1}^n-\beta u_{j+1}^n+\beta u_{j-1}^n+\gamma s_x^n
\end{aligned}
$$

y de manera similar para $v$:
<br>
$$
\begin{aligned}
v_{i j}^{n+1}=v_{i, j}^n & -\alpha v_{i+1}^{n+1}+\alpha v_{i-1}^{n+1}-\beta v_{j+1}^{n+1}+\beta v_{j-1}^{n+1}+\gamma s_y^{n+1} \\
& -\alpha v_{i+1}^n+\alpha v_{i-1}^n-\beta v_{j+1}^n+\beta v_{j-1}^n+\gamma s_y^n
\end{aligned}
$$
