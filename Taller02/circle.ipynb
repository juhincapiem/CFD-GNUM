```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def divisions(ratio, r):
  Delta1 = np.pi*ratio[0]
  Delta2 = np.pi*ratio[1]
  Delta3 = np.pi*ratio[2]

  points = 20
  x = np.zeros((points, 1))
  y = np.zeros((points, 1))

  semiCircle = np.linspace(0, np.pi, points, endpoint=True)
  x[:, 0] = r*np.cos(semiCircle)
  y[:, 0] = r*np.sin(semiCircle)

  xCoord = np.array([np.cos(Delta1), np.cos(Delta1+Delta2)])*r
  yCoord = np.array([np.sin(Delta1), np.sin(Delta1+Delta2)])*r

  fig, ax = plt.subplots()
  ax.plot(x, y, "--r")
  for i in range(0, len(xCoord)):
    ax.plot([xCoord[i],0], [yCoord[i],0], "k-")
    ax.plot([xCoord[i],0], [yCoord[i],0], "ok")
  plt.axis("equal")
  ax.grid("True")
  plt.show()
  return x, y


def upperCircle(ratio, r, points, x1, y1):
  Delta = np.array([0, ratio[2], ratio[1], ratio[2]])*np.pi
  fig, ax = plt.subplots()
  for i in range(3):
    semiCircle = np.linspace(sum(Delta[0:i+1]), sum(Delta[0:i+2]), points[2], endpoint=True)

    coordinates = np.zeros((points[i], 2))
    coordinates[:, 0] = r*np.cos(semiCircle)+x1
    coordinates[:, 1] = r*np.sin(semiCircle)+y1

    ax.plot(coordinates[:,0], coordinates[:,1], "or")

    if i == 0:
      with open('upperCircle.txt', 'w') as interpol:
        interpol.write('delta {}\n'.format(i+1))
        for x, y in coordinates:
          interpol.write('({:.4f} {:.4f} 0.5)\n'.format(x, y))
        interpol.close

    else:
        with open('upperCircle.txt', 'a') as interpol:
          interpol.write('delta {}\n'.format(i+1))
          for x, y in coordinates:
              interpol.write('({:.4f} {:.4f} 0.5)\n'.format(x, y))
          interpol.close
  plt.axis("equal")
  plt.show()

def bottomCircle(ratio, r, points, x1, y1):
  Delta = np.array([0, -ratio[2], -ratio[1], -ratio[2]])*np.pi

  fig, ax = plt.subplots()
  for i in range(3):
    semiCircle = np.linspace(sum(Delta[0:i+1]), sum(Delta[0:i+2]), points[2], endpoint=True)

    coordinates = np.zeros((points[i], 2))
    coordinates[:, 0] = r*np.cos(semiCircle)+x1
    coordinates[:, 1] = r*np.sin(semiCircle)+y1

    ax.plot(coordinates[:,0], coordinates[:,1], "or")

    if i == 0:
      with open('bottomCircle.txt', 'w') as interpol:
        interpol.write('delta {}\n'.format(i+1))
        for x, y in coordinates:
          interpol.write('({:.4f} {:.4f} -0.5)\n'.format(x, y))
        interpol.close

    else:
        with open('bottomCircle.txt', 'a') as interpol:
          interpol.write('delta {}\n'.format(i+1))
          for x, y in coordinates:
              interpol.write('({:.4f} {:.4f} -0.5)\n'.format(x, y))
          interpol.close
  plt.axis("equal")
  plt.show()
```

```python
#Da치metro
D = 2
#Radio
r = D/2
#Proporci칩n
ratio = np.array([0.3, 0.4, 0.3])

#Delta de distancia en direcci칩n X
x1 = 2*D
x2 = 20*D

#Delta de distancia en direcci칩n Y
y1 = 2*D
y2 = 2.1*D

x, y = divisions(ratio, r)
print(x.transpose, y.transpose)
```

```python
points = np.array([5,5,5])
upperCircle(ratio, r, points, x1, y1)
bottomCircle(ratio, r, points, x1, y1)
```
